Я создам фильтр, который будет применять размытие к изображению. Это может быть полезно для создания эффекта мягкости или уменьшения деталей.

```python
import requests
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
from io import BytesIO

def apply_blur_filter(image, radius=5):
    """
    Применяет размытие к изображению.

    :param image: Входное изображение (PIL Image)
    :param radius: Радиус размытия
    :return: Размытое изображение
    """
    return image.filter(ImageFilter.GaussianBlur(radius))

# Пример использования
if __name__ == "__main__":
    # URL изображения
    url = "https://i.pinimg.com/736x/7c/cd/09/7ccd09e2648f10d2c81cbf2c19709bfd.jpg"

    # Загружаем изображение
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    # Применяем встроенные возможности Pillow
    resized_img = img.resize((300, 300))  # Изменение размера
    cropped_img = img.crop((50, 50, 250, 250))  # Обрезка
    blurred_img = apply_blur_filter(img, radius=10)  # Применение нашего фильтра

    # Отображаем оригинальное и измененные изображения
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.title("Оригинальное изображение")
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.title("Измененное изображение (размер)")
    plt.imshow(resized_img)
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.title("Обрезанное изображение")
    plt.imshow(cropped_img)
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.title("Размытое изображение")
    plt.imshow(blurred_img)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
